```python
import yfinance as yf
import pandas as pd
```

```python
symbols = [
'X',
'GM',
'AAPL',
'AMZN',
'META',
'GOOG',
"TTD",
"DELL",
'IBM',
'MSFT',
'NET',
'NFLX',
'TSLA',
'HD',
"LULU",
]
```

```python
companies =  []
for s in symbols:
    stock = yf.Ticker(s)
    print(s)
    record = create_record(stock)
    companies.append(record)
```

```python
def create_record(stock):
    sd = {}
    sd["symbol"] = stock.info["symbol"]
    sd['name'] = stock.info.get("shortName", None)
    sd["exchange"] = stock.info["exchange"]
    sd['industry'] = stock.info["industry"]
    sd['sector'] = stock.info["sector"]
    sd['info'] = {}
    sd['info']['website'] = stock.info["website"]
    sd['info']['city'] = stock.info["state"]
    sd['info']['state'] = stock.info["city"]
    sd['info']['country'] = stock.info["country"]
    return sd
```

```python
import json 
with open("companies.json", "w") as f:
    json.dump(companies, f)
```

```python
for s in symbols:
    stock = yf.Ticker(s)
    sh = stock.history(period="1wk")
    sh["Date"] = sh.index
    sh["Date"] = sh['Date'].apply(str)
    sh['Date'] = sh['Date'].str[0:10]
    sh["Symbol"] = s
    export = sh[["Date","Symbol","Open","High","Low","Close","Volume"]]
    export.to_csv(f"{s}.csv", index=False, header=True)
```
